# 統計學簡述

統計學是一門利用數據進行收集、整理、分析和解釋的學科，旨在幫助人們從數據中提取有意義的信息。
統計學主要分為以下兩大分支：

1. **描述統計 (Descriptive Statistics)**:
   - 用於總結和描述數據的特徵。
   - 包括：
     - **集中趨勢測量**：平均數 (Mean)、中位數 (Median)、眾數 (Mode)。
     - **離散程度測量**：範圍 (Range)、四分位距 (IQR)、方差 (Variance)、標準差 (Standard Deviation)。
     - **圖形表示**：直方圖 (Histogram)、餅圖 (Pie Chart)、箱型圖 (Box Plot) 等。

2. **推論統計 (Inferential Statistics)**:
   - 用於從樣本數據中推斷總體特徵。
   - 包括：
     - **點估計與區間估計**：估計總體參數（如平均數）。
     - **假設檢定**：t 檢定、卡方檢定、ANOVA。
     - **相關性與迴歸分析**：分析變數之間的關係。

# 常用方法與應用

1. **數據收集**：
   - 問卷調查、實驗數據、觀察數據。

2. **數據分析**：
   - 探索性數據分析 (EDA)：檢查數據的基本結構與分佈。
   - 模型建構：利用統計模型解釋數據行為。

3. **應用領域**：
   - 醫學：臨床試驗、流行病學研究。
   - 商業：市場分析、風險評估。
   - 社會科學：人口統計、行為研究。

# 1. 描述性統計 (Descriptive Statistics)
- 用於總結和描述數據的特徵。
  - 平均數 (Mean)：數據的集中趨勢。
  - 中位數 (Median)：將數據分成兩部分的中間值。
  - 標準差 (Standard Deviation)：衡量數據的分散程度。
  - 直方圖 (Histogram)：展示數據分佈的圖形。

# 2. 假設檢定 (Hypothesis Testing)
- 用於判斷數據是否支持某個假設。
  - t 檢定 (t-test)：比較兩組數據的均值。
    - 單樣本 t 檢定：檢驗樣本均值是否與特定值相等。
    - 成對樣本 t 檢定：比較同一組樣本在不同條件下的均值。
    - 獨立樣本 t 檢定：比較兩組獨立樣本的均值。
  - 卡方檢定 (Chi-square Test)：檢查分類數據的分佈是否符合預期。
    - 獨立性檢定：測試兩個變數是否相關。
    - 擬合優度檢定：檢驗觀測值是否符合理論分佈。
  - ANOVA (分析變異數)：檢測三個或更多組的均值是否有顯著差異。

# 3. 相關性分析 (Correlation Analysis)
- 衡量兩個變數之間的關聯程度。
  - 皮爾森相關係數 (Pearson Correlation Coefficient)：測試線性相關性。
  - 斯皮爾曼相關係數 (Spearman Rank Correlation)：測試秩次相關性，對非線性數據更適用。

# 4. 回歸分析 (Regression Analysis)
- 用於研究變數之間的依賴關係。
  - 線性回歸 (Linear Regression)：檢查一個自變數與應變數之間的線性關係。
  - 多元線性回歸 (Multiple Linear Regression)：檢查多個自變數與應變數之間的關係。
  - 邏輯回歸 (Logistic Regression)：用於處理二元或多分類數據。

# 5. 聚類分析 (Clustering)
- 將數據分組，以識別具有相似特徵的群體。
  - K 均值聚類 (K-Means Clustering)。
  - 階層聚類 (Hierarchical Clustering)。

# 6. 主成分分析 (PCA, Principal Component Analysis)
- 減少數據維度，同時保留盡可能多的信息。

# 7. 生存分析 (Survival Analysis)
- 研究時間到事件的分佈。
  - Kaplan-Meier 曲線：用於生存概率的估計。
  - Cox 比例風險模型：研究多個變數對生存時間的影響。

# 8. 時間序列分析 (Time Series Analysis)
- 分析數據隨時間變化的趨勢和模式。
  - 自迴歸模型 (AR)。
  - 移動平均模型 (MA)。
  - ARIMA 模型：結合 AR 和 MA，用於預測。

# 9. 數據分佈檢測
- 檢測數據是否符合某種分佈。
  - Shapiro-Wilk 檢定：檢測數據是否為正態分佈。
  - Kolmogorov-Smirnov 檢定：檢驗數據分佈與參考分佈的相符程度。

# 10. 數據降維
- 用於處理高維數據，降低計算複雜度。
  - 主成分分析 (PCA)。
  - 奇異值分解 (SVD)。

# 常見應用領域
1. **醫學研究**：臨床試驗、疾病風險評估。
2. **市場分析**：顧客行為建模、銷售預測。
3. **社會科學**：人口統計、心理測量。
4. **金融分析**：投資風險評估、股票價格預測。


```python
# 必要模組
import numpy as np
from scipy import stats

# 範例 1: 單樣本 t 檢定 (One-sample t-test)
data = [98, 100, 102, 97, 101]
t_stat, p_value = stats.ttest_1samp(data, 100)
print(f"範例 1: 單樣本 t 檢定結果\n t 統計量: {t_stat}, p 值: {p_value}")
```

```python
# 範例 2: 成對樣本 t 檢定 (Paired t-test)
before = [88, 90, 92, 85, 87]
after = [90, 91, 93, 88, 89]
t_stat, p_value = stats.ttest_rel(before, after)
print(f"範例 2: 成對樣本 t 檢定結果\n t 統計量: {t_stat}, p 值: {p_value}")
```

```python
# 範例 3: 獨立樣本 t 檢定 (Independent t-test)
group1 = [85, 87, 89, 90, 92]
group2 = [88, 86, 84, 87, 89]
t_stat, p_value = stats.ttest_ind(group1, group2)
print(f"範例 3: 獨立樣本 t 檢定結果\n t 統計量: {t_stat}, p 值: {p_value}")
```

```python
# 範例 4: 卡方檢定 (Chi-square Test)
observed = [25, 35, 40]
expected = [30, 30, 40]
chi2, p_value = stats.chisquare(f_obs=observed, f_exp=expected)
print(f"範例 4: 卡方檢定結果\n Chi2 統計量: {chi2}, p 值: {p_value}")
```

```python
# 範例 5: 列聯表的卡方檢定 (Chi-square Contingency Test)
table = [[50, 30], [20, 40]]
chi2, p, dof, expected = stats.chi2_contingency(table)
print(f"範例 5: 列聯表卡方檢定結果\n Chi2 統計量: {chi2}, p 值: {p}, 自由度: {dof}")
```

```python
# 範例 6: 單因子變異數分析 (One-way ANOVA)
group1 = [88, 90, 92, 85, 87]
group2 = [90, 91, 93, 88, 89]
group3 = [95, 96, 94, 97, 93]
f_stat, p_value = stats.f_oneway(group1, group2, group3)
print(f"範例 6: 單因子變異數分析結果\n F 統計量: {f_stat}, p 值: {p_value}")
```

```python
# 範例 7: 皮爾森相關檢定 (Pearson Correlation Test)
x = [10, 20, 30, 40, 50]
y = [15, 25, 35, 45, 55]
corr, p_value = stats.pearsonr(x, y)
print(f"範例 7: 皮爾森相關檢定結果\n 相關係數: {corr}, p 值: {p_value}")
```

```python
# 範例 8: 斯皮爾曼相關檢定 (Spearman Correlation Test)
x = [10, 20, 30, 40, 50]
y = [5, 15, 25, 35, 65]
corr, p_value = stats.spearmanr(x, y)
print(f"範例 8: 斯皮爾曼相關檢定結果\n Spearman 相關係數: {corr}, p 值: {p_value}")
```

```python
# 範例 9: Shapiro-Wilk 正態性檢定 (Normality Test)
data = [85, 87, 89, 90, 92]
stat, p_value = stats.shapiro(data)
print(f"範例 9: Shapiro-Wilk 正態性檢定結果\n 統計量: {stat}, p 值: {p_value}")
```

```python
# 範例 10: Kolmogorov-Smirnov 檢定 (Kolmogorov-Smirnov Test)
data = np.random.normal(loc=0, scale=1, size=100)
stat, p_value = stats.kstest(data, 'norm')
print(f"範例 10: Kolmogorov-Smirnov 檢定結果\n K-S 統計量: {stat}, p 值: {p_value}")
```

```python
# 範例 11: Mann-Whitney U 檢定 (Mann-Whitney U Test)
group1 = [88, 90, 92, 85, 87]
group2 = [90, 91, 93, 88, 89]
u_stat, p_value = stats.mannwhitneyu(group1, group2)
print(f"範例 11: Mann-Whitney U 檢定結果\n U 統計量: {u_stat}, p 值: {p_value}")
```

# 降維方法簡介：PCA、MDS、UMAP 和 t-SNE

降維方法常用於將高維數據降到較低維度，方便可視化、降噪或預處理。以下簡單介紹常見的降維方法：

---

## 📊 1. 主成分分析 (PCA)

- **類型：** 線性降維  
- **目的：** 保留數據最多的變異性。  
- **優點：**
  - 計算速度快  
  - 簡單高效  
- **缺點：**
  - 只適用於線性數據  
- **使用場景：** 預處理、降噪、降維。

---

## 🔍 2. 多維尺度分析 (MDS)

- **類型：** 線性和非線性混合  
- **目的：** 保留樣本之間的距離結構。  
- **優點：**  
  - 可以保留數據間的距離信息。  
- **缺點：**
  - 計算較慢，適用性受數據規模影響。  
- **使用場景：** 距離結構保留。

---

## 🔧 3. UMAP (Uniform Manifold Approximation and Projection)

- **類型：** 非線性降維  
- **目的：** 更好地保留局部和全局數據結構。  
- **優點：**
  - 計算速度快  
  - 保留非線性結構效果好。  
- **使用場景：** 大數據可視化、聚類預處理。

---

## 🔎 4. t-SNE (t-distributed Stochastic Neighbor Embedding)

- **類型：** 非線性降維  
- **目的：** 在可視化高維數據時保留局部結構。  
- **優點：**
  - 在可視化上效果出色。  
- **缺點：**
  - 計算速度較慢。  
- **使用場景：** 主要用於數據可視化。

---

## 📊 **比較總結：**

| **方法**  | **類型**  | **保留結構**       | **計算速度**     | **使用場景**         |
|------------|----------|--------------------|------------------|----------------------|
| **PCA**     | 線性     | 保留變異性       | 很快             | 預處理、降噪        |
| **MDS**     | 混合     | 保留距離結構     | 慢               | 距離保留            |
| **UMAP**    | 非線性   | 保留局部與全局結構 | 很快             | 可視化、聚類        |
| **t-SNE**    | 非線性   | 保留局部結構     | 非常慢           | 數據可視化         |

---

## 📌 **選擇建議：**

1. **PCA：** 適合高維預處理，速度快。  
2. **MDS：** 適用於保留距離結構。  
3. **UMAP：** 更適合大數據集和非線性結構保留。  
4. **t-SNE：** 適合數據可視化，但計算較慢。

根據數據場景和需求選擇適合的方法！📊


```python
# 必要模組
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import MDS, TSNE
import umap

# 隨機生成數據
np.random.seed(42)
data = np.random.rand(100, 10)  # 100 個樣本，每個有 10 個特徵
```

```python
# PCA 降維到 2 維
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data)

# 視覺化
plt.scatter(data_pca[:, 0], data_pca[:, 1], c='blue', alpha=0.7)
plt.title("PCA Projection")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

```python
# MDS 降維到 2 維
mds = MDS(n_components=2, random_state=42)
data_mds = mds.fit_transform(data)

# 視覺化
plt.scatter(data_mds[:, 0], data_mds[:, 1], c='green', alpha=0.7)
plt.title("MDS Projection")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.show()
```

```python
# UMAP 降維到 2 維
umap_model = umap.UMAP(n_components=2, random_state=42)
data_umap = umap_model.fit_transform(data)

# 視覺化
plt.scatter(data_umap[:, 0], data_umap[:, 1], c='red', alpha=0.7)
plt.title("UMAP Projection")
plt.xlabel("UMAP Dimension 1")
plt.ylabel("UMAP Dimension 2")
plt.show()
```

```python
# t-SNE 降維到 2 維
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
data_tsne = tsne.fit_transform(data)

# 視覺化
plt.scatter(data_tsne[:, 0], data_tsne[:, 1], c='purple', alpha=0.7)
plt.title("t-SNE Projection")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.show()
```
